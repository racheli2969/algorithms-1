**×”× ×—×ª ×”××™× ×“×•×§×¦×™×”:**

$$\sum_{i=0}^{n} 2^i = 2^{n+1} - 1$$

**×¦×¢×“ ×”××™× ×“×•×§×¦×™×”:**

× ×•×›×™×— ×©×”× ×•×¡×—×” × ×›×•× ×” ×¢×‘×•×¨ $k = n + 1$:

$$\sum_{i=0}^{n+1} 2^i = \sum_{i=0}^{n} 2^i + 2^{n+1}$$

×œ×¤×™ ×”× ×—×ª ×”××™× ×“×•×§×¦×™×”:

$$= (2^{n+1} - 1) + 2^{n+1}$$

×•×–×” ×‘×“×™×•×§ ××” ×©×¨×¦×™× ×• ×œ×”×•×›×™×— ×¢×‘×•×¨ $k = n + 1$.

## ×—×œ×§ ×' - ××™××•×© ×”×¤×•× ×§×¦×™×” find_min_max

×›×ª×‘×™ ×¤×•× ×§×¦×™×” ×‘-Python ×©××§×‘×œ×ª ××¢×¨×š/×¨×©×™××” ×¢× ×¢×¨×›×™× ×•××—×–×™×¨×” ××ª ×”××™× ×™××•× ×•×”××§×¡×™××•×.

**×“×¨×™×©×•×ª:**
- ××™×Ÿ ×œ×”×©×ª××© ×‘×¤×•× ×§×¦×™×•×ª `min` ××• `max` ×”××•×‘× ×•×ª ×‘-Python
- ×—×ª×™××ª ×”×¤×•× ×§×¦×™×”: `find_min_max(a, key)` 
- ×”×¤×¨××˜×¨ `key` ×”×•× ×¤×•× ×§×¦×™×” ×©××—×–×™×¨×” ××¤×ª×— ×©×œ×¤×™×• ××©×•×•×™× ××ª ×”×¤×¨×™×˜×™×

````python
from typing import List, Any, Callable, Tuple
import random
import string

def find_min_max(a: List[Any], key: Callable[[Any], Any]) -> Tuple[Any, Any]:
    """
    ××•×¦××ª ××ª ×”××™× ×™××•× ×•×”××§×¡×™××•× ×‘×¨×©×™××” ×œ×¤×™ ×¤×•× ×§×¦×™×™×ª ×”××¤×ª×—.
    
    Parameters:
        a: ×¨×©×™××” ×©×œ ×¤×¨×™×˜×™×
        key: ×¤×•× ×§×¦×™×” ×©××—×–×™×¨×” ×¢×¨×š ×œ×”×©×•×•××” ××›×œ ×¤×¨×™×˜
    
    Returns:
        tuple: (×¤×¨×™×˜_××™× ×™××œ×™, ×¤×¨×™×˜_××§×¡×™××œ×™)
    
    Raises:
        ValueError: ×× ×”×¨×©×™××” ×¨×™×§×”
    """
    if not a:
        raise ValueError('×”×¨×©×™××” ×¨×™×§×”')
    
    # ××ª×—×•×œ - ×”×¤×¨×™×˜ ×”×¨××©×•×Ÿ ×”×•× ×’× ×”××™× ×™××•× ×•×’× ×”××§×¡×™××•× ×”×”×ª×—×œ×ª×™×™×
    min_item = a[0]
    max_item = a[0]
    min_key = key(min_item)
    max_key = key(max_item)
    
    # ×¢×•×‘×¨ ×¢×œ ×©××¨ ×”×¤×¨×™×˜×™× ×‘×¨×©×™××”
    for item in a[1:]:
        current_key = key(item)
        
        # ×‘×“×™×§×” ×× × ××¦× ××™× ×™××•× ×—×“×©
        if current_key < min_key:
            min_key = current_key
            min_item = item
        
        # ×‘×“×™×§×” ×× × ××¦× ××§×¡×™××•× ×—×“×©
        if current_key > max_key:
            max_key = current_key
            max_item = item
    
    return min_item, max_item


# ×‘×“×™×§×ª ×”×¤×•× ×§×¦×™×” ×¢× ×“×•×’××” ×¤×©×•×˜×”
test_data = [(3, 'c'), (1, 'a'), (5, 'e'), (2, 'b')]
min_item, max_item = find_min_max(test_data, key=lambda x: x[0])
print(f"×“×•×’××”: {test_data}")
print(f"××™× ×™××•× ×œ×¤×™ ×¤×¨×™×˜ ×¨××©×•×Ÿ: {min_item}")
print(f"××§×¡×™××•× ×œ×¤×™ ×¤×¨×™×˜ ×¨××©×•×Ÿ: {max_item}")
````

### ×¤×•× ×§×¦×™×” ×¢×–×¨: create_random_tuples

×œ×¤× ×™ ×©× ××©×™×š ×œ×—×œ×§ ×‘', × ×××© ××ª ×”×¤×•× ×§×¦×™×” `create_random_tuples` ×©×ª×©××© ××•×ª× ×• ×œ×™×¦×™×¨×ª × ×ª×•× ×™ ×‘×“×™×§×”:

````python
def create_random_tuples(n: int, k: int, types: List[type]) -> List[Tuple]:
    """
    ×™×•×¦×¨×ª n tuples, ×›×œ ××—×“ ×¢× k ×¤×¨×™×˜×™× ××”×¡×•×’×™× ×”××‘×•×§×©×™×.
    
    Parameters:
        n: ××¡×¤×¨ ×”-tuples ×œ×™×¦×™×¨×”
        k: ××¡×¤×¨ ×”×¤×¨×™×˜×™× ×‘×›×œ tuple
        types: ×¨×©×™××ª ×˜×™×¤×•×¡×™× (int, float, str)
    
    Returns:
        List[Tuple]: ×¨×©×™××” ×©×œ tuples ××§×¨××™×™×
    
    ×”×¢×¨×”:
        - int: ××¡×¤×¨ ×©×œ× ××§×¨××™ ×‘×™×Ÿ 0 ×œ-1000
        - float: ××¡×¤×¨ ×¢×©×¨×•× ×™ ××§×¨××™ ×‘×™×Ÿ 0 ×œ-1000
        - str: ××—×¨×•×–×ª ×‘×ª 5 ×ª×•×•×™× ××§×¨××™×™×
    """
    result = []
    
    for _ in range(n):
        tuple_items = []
        
        for i in range(k):
            # ×‘×•×—×¨ ××ª ×”×˜×™×¤×•×¡ ×”××ª××™× ×œ×¤×™ ×”××™×§×•×
            current_type = types[i % len(types)]
            
            if current_type is int:
                # ××¡×¤×¨ ×©×œ× ××§×¨××™
                tuple_items.append(random.randint(0, 1000))
            elif current_type is float:
                # ××¡×¤×¨ ×¢×©×¨×•× ×™ ××§×¨××™
                tuple_items.append(random.random() * 1000)
            elif current_type is str:
                # ××—×¨×•×–×ª ××§×¨××™×ª ×‘××•×¨×š 5
                tuple_items.append(''.join(random.choices(string.ascii_letters + string.digits, k=5)))
            else:
                # ×‘×¨×™×¨×ª ××—×“×œ - × ×™×¡×™×•×Ÿ ×œ×§×¨×•× ×œ×‘× ××™
                tuple_items.append(current_type())
        
        result.append(tuple(tuple_items))
    
    return result


# ×“×•×’××” ×œ×©×™××•×©
sample_tuples = create_random_tuples(5, 3, [int, float, str])
print("5 tuples ××§×¨××™×™× ×¢× 3 ×¤×¨×™×˜×™× ××¡×•×’ [int, float, str]:\n")
for i, t in enumerate(sample_tuples, 1):
    print(f"{i}. {t}")
````

- ×‘×›×œ ×¤×¢× ××©×ª××©×™× ×‘×¤×¨×™×˜ ××—×¨ ×©×œ ×”-tuple ×›××¤×ª×— ×”××™×•×Ÿ

## ××™××•×© Insertion Sort

````python
def insertion_sort(a: List[Any], key: Callable[[Any], Any]) -> None:
    """
    ×××™×™×Ÿ ×¨×©×™××” ×‘××§×•× (in-place) ×‘×××¦×¢×•×ª ××œ×’×•×¨×™×ª× insertion sort.
    
    Parameters:
        a: ×”×¨×©×™××” ×œ××™×•×Ÿ (×ª×©×•× ×” ×‘××§×•×)
        key: ×¤×•× ×§×¦×™×” ×©××—×–×™×¨×” ×¢×¨×š ×œ×”×©×•×•××” ××›×œ ×¤×¨×™×˜
    
    Returns:
        None (×”×¨×©×™××” ××©×ª× ×” ×‘××§×•×)
    
    ××œ×’×•×¨×™×ª×:
        ×¢×‘×•×¨ ×›×œ ××œ×× ×˜ ××”×©× ×™ ×•×”×œ××”:
        1. ×©×•××¨×™× ××ª ×”××œ×× ×˜ ×”× ×•×›×—×™
        2. ××©×•×•×™× ××•×ª×• ×œ××œ×× ×˜×™× ×œ×¤× ×™×• (×©×›×‘×¨ ×××•×™× ×™×)
        3. ××–×™×–×™× ××œ×× ×˜×™× ×’×“×•×œ×™× ×™×•×ª×¨ ××§×•× ××—×“ ×™××™× ×”
        4. ××›× ×™×¡×™× ××ª ×”××œ×× ×˜ ×”× ×•×›×—×™ ×œ××§×•× ×”× ×›×•×Ÿ
    """
    # ×¢×•×‘×¨ ×¢×œ ×›×œ ××œ×× ×˜ ×”×—×œ ××”×©× ×™ (××™× ×“×§×¡ 1)
    for j in range(1, len(a)):
        # ×©×•××¨ ××ª ×”××œ×× ×˜ ×”× ×•×›×—×™ ×•××ª ×¢×¨×š ×”××¤×ª×— ×©×œ×•
        current = a[j]
        current_key = key(current)
        
        # ××ª×—×™×œ ×œ×”×©×•×•×ª ×¢× ×”××œ×× ×˜ ×©×œ×¤× ×™×•
        i = j - 1
        
        # ××–×™×– ××œ×× ×˜×™× ×©×’×“×•×œ×™× ××”× ×•×›×—×™ ××§×•× ××—×“ ×™××™× ×”
        while i >= 0 and key(a[i]) > current_key:
            a[i + 1] = a[i]
            i -= 1
        
        # ××›× ×™×¡ ××ª ×”××œ×× ×˜ ×”× ×•×›×—×™ ×œ××§×•× ×”× ×›×•×Ÿ
        a[i + 1] = current


# ×‘×“×™×§×” ×¤×©×•×˜×” ×©×œ ×”××œ×’×•×¨×™×ª×
test_list = [5, 2, 8, 1, 9, 3]
print(f"×œ×¤× ×™ ××™×•×Ÿ: {test_list}")
insertion_sort(test_list, key=lambda x: x)
print(f"××—×¨×™ ××™×•×Ÿ: {test_list}")
````
